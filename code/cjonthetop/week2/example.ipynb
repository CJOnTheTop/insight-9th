### GroupBy() 적용

```python
data_groupby = data.groupby(by="pclass")
print(type(data_groupby))
```

#### DataFrameGroupBy 객체에 [["pclass", "survived"]]로 필터링해서 count() 하기

```python
data_groupby = data.groupby(by="pclass")[["pclass","survived"]].count()
data_groupby
```

```python
#agg()를 이용해 여러 함수 구현도 가능
agg_format = {"age":"max", "sibsp":"sum", "fare":"mean"}
data.groupby("pclass").agg(agg_format)
```
